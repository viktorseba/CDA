## Imports

```python
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import matplotlib
from pathlib import Path
import time
# %matplotlib notebook
```

## Load dataset

```python
path = Path().cwd()
fiel_path = path.parent / "Data" / "simulatedData.csv"
X = np.loadtxt(fiel_path, delimiter = ",")
```

## Helper Functions

```python
def plotting(X, kmeans, y_pred, ax):
    ax.clear()
    ax.scatter(X[:,0], X[:,1], c = y_pred, marker= ".", alpha=0.5)
    ax.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], c = np.arange(kmeans.cluster_centers_.shape[0]), marker= "*", s=100)
    ax.set_title("Current iteration")
```

```python
clusters = 3 # Control how many clusters we want
   
from time import sleep
# impor clear_output
from IPython.display import clear_output
plt.suptitle("K-means Demo")

fig, ax = plt.subplots(figsize = (16,10))
kmeans = KMeans(n_clusters = clusters, n_init = 1, random_state = 1, max_iter = 1)
y_pred = kmeans.fit_predict(X)

plt.scatter(X[:,0], X[:,1], marker = '.', c = y_pred,animated = True)
plt.show(block=False)
sleep(0.5)
clear_output(wait=True)

current_iter = 0
total_iter = 20


for _ in range(total_iter):
   
    fig, ax = plt.subplots(figsize = (12,8))
    kmeans = KMeans(n_clusters = clusters, init = "random", n_init = 1, random_state = 0, max_iter = 1 + _)
    y_pred = kmeans.fit_predict(X)
   
    plt.scatter(X[:,0], X[:,1], c = y_pred, marker= ".",animated = True)
    plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], c = np.arange(kmeans.cluster_centers_.shape[0]), marker= "*", s=200,animated = True,linewidths=.2,edgecolors="black")
    plt.title(f"Current iteration {_}")
    plt.show(block=False)
   
   
    sleep(0.5)
    clear_output(wait=True)
```
